### Updating cells using loc and iloc

```python
import pandas as pd

column_values = [f'Column{i}' for i in range(1, 5)]
index_values = [chr(i) for i in range(65, 72)]

df = pd.DataFrame([[f'{ind}{col}' for col in range(1, 5)] for ind in index_values], 
                  index=index_values, columns=column_values)

df
```

```python
copy = df.copy()
copy.iloc[1, 2] = 'xx'

copy
```

```python
# rows and columns can be added using loc but not iloc

copy = df.copy()
copy.loc['H', :] = ['H1', 'H2', 'H3', 'H4']
copy.loc[:, 'Column5'] = 0
copy.loc[:, 'Column6'] = copy.loc[:, 'Column1'] + '-' + copy.loc[:, 'Column2']
copy.loc[:, 'Column7'] = copy.loc[:, 'Column6'] # smidigare sätt än raden ovanför ^



copy
```

### Renaming columns and indexes

```python
copy = df.copy()
#list(copy.columns)
copy.columns = ['A', 'B', 'C', 'D']
copy.rename(columns={'B': 'ColumnB', 'D': 'Johan'}, inplace=True)

copy
```

### Delete reows and columns

```python
copy = df.copy()
copy.drop(index=['D', 'F'], columns='Column3', inplace=True)

copy
```
