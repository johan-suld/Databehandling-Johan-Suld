#### Missing values

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
df = df[df > 0]

df
```

```python
copy = df.copy()
copy.drop(columns='D', inplace=True)

#copy
```

```python
#copy.isna()

copy.dropna() # droppar alla rander som innehåller minst en NaN
copy.dropna(how='all') # droppar alla rader som endast innehåller NaN
copy.dropna(thresh=3) # måste innehålla tre värden för att vara kvar
copy.dropna(subset=['A', 'C']) # letar endast efter NaN i A och C
copy.dropna(axis='columns', thresh=5) # droppar kolumner istället för rader
```

```python
copy = df.copy()
copy.fillna(0)

copy['A'].fillna(copy['A'].mean(), inplace=True)
#copy['B'].mode()[0]
copy['B'].ffill(inplace=True)
copy['D'].bfill(inplace=True)
copy['C'].interpolate(inplace=True)

copy
```
